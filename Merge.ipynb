Alright, data is loaded up, now let's do some preliminary cleaning so we can merge.

The biggest pain is going to be dealing with the FBI crime data. It doesn't appear as if it has any CBSA code associated with it, so we're going to have to match on the metro area name after we do some hacky string manipulation. It's going to really suck to manage, but it's gotta get done.

Also our definition of 'excessive drinking' is fuzzy. We have a few different questions we can draw from: 
- 'Heavy drinkers (adult men having more than 14 drinks per week and adult women having more than 7 drinks per week) (variable calculated from one or more BRFSS questions)'
- 'Heavy drinkers (adult men having more than two drinks per day and adult women having more than one drink per day) (variable calculated from one or more BRFSS questions)'
- 'Binge drinkers (males having five or more drinks on one occasion, females having four or more drinks on one occasion) (variable calculated from one or more BRFSS questions)'
- 'Respondents who have reported having driven after drinking too much (variable calculated from one or more BRFSS questions)'

For smoking, we'll use this question:
- 'Adults who are current smokers (variable calculated from one or more BRFSS questions)'

Here's another problem: the population statistics proided by the Census Bureau aren't even close to the ones offered by the CDC in their accidental death data. Which set should we use?

```python
#Cleaning accidental death data
accidental_death = accidental_death.iloc[:188,1:]
accidental_death = accidental_death.drop(['Crude Rate', 'Deaths'],axis=1)
accidental_death.rename(columns={
    '2013 CBSA Code' : 'CBSA Code',
    'Age Adjusted Rate' : 'Age Adjusted Accidental Death Rate per 100000'
}, inplace=True)
```

```python
#Clean air quality data
airq = airq[['CBSA Code', 'Median AQI']]
```

```python
#Clean population data
population = population[['Id2', 'Estimate; Total']]
population.rename(columns={
    'Id2' : 'CBSA Code'
}, inplace=True)
```

```python
#Extract the relevant questions from the smoking_drinking dataset
#using the most recent year available;
#construct new columns and append to the dataframe.
smoking = smoking_drinking.loc[
    (smoking_drinking['Question'] == 'Adults who are current smokers (variable calculated from one or more BRFSS questions)') &
    (smoking_drinking['Response'] == 'Yes')
][['Year', 'Locationabbr', 'Data_value']]
smoking.rename(columns={
    'Locationabbr' : 'CBSA Code',
    'Data_value' : 'Smoking rate'
},inplace=True)

#De-dupe by location code, most recent year
for ID in smoking['CBSA Code'].unique():
    max_year = max(smoking.loc[
        smoking['CBSA Code'] == ID
    ]['Year'])
    smoking.drop(smoking.loc[
        (smoking['CBSA Code'] == ID) &
        ~(smoking['Year'] == max_year)
    ].index, axis=0, inplace=True)
    
drinking = smoking_drinking.loc[
    (smoking_drinking['Question'] == 'Binge drinkers (males having five or more drinks on one occasion, females having four or more drinks on one occasion) (variable calculated from one or more BRFSS questions)') &
    (smoking_drinking['Response'] == 'Yes')
][['Year', 'Locationabbr', 'Data_value']]
drinking.rename(columns={
    'Locationabbr' : 'CBSA Code',
    'Data_value' : 'Binge drinking rate'
},inplace=True)

for ID in drinking['CBSA Code'].unique():
    max_year = max(drinking.loc[
        drinking['CBSA Code'] == ID
    ]['Year'])
    drinking.drop(drinking.loc[
        (drinking['CBSA Code'] == ID) &
        ~(drinking['Year'] == max_year)
    ].index, axis=0, inplace=True)
    
drinking.drop(columns=['Year'], axis=1,inplace=True)
```

```python
#Merge our datasets
DATASETS = [accidental_death, airq, population, smoking, drinking]
merged = reduce(lambda x,y: x.merge(y, on='CBSA Code', how='inner'), DATASETS)
```

```python
'''
Alright, time for the ugly part.

To extract all of the FBI data, we're going to have to:
    1) Find every row that doesn't have NaN for Metropolitan Statistical Area
    2) Add 4 to the index to get the row with 'Rate per 100,000 inhabitants' under 'Counties/principal cities'
        (Side note: whose bright idea was that?)
    3) Extract the 'Violent Crime' and 'Property Crime' values, if they exist
        Create a corner case for when they don't!
    4) Pull the 'Metropolitan Statistical Area' string and extract the location in the format <METRO>, <STATE>
    5) Match that against the string we have under '2013 CBSA' in the merged table'

...simple enough, right?
'''

#First, cut down the number of rows in the crime dataset
crime = crime.iloc[:2331,:]

#Extract the indices of rows that have the metro name
metro_indices = crime[pd.notnull(crime['Metropolitan Statistical Area'])].index

#Find the rate per 100,000 for each metro area
final_metro_indices = []
violent_crime_rate = []
property_crime_rate = []
for x in range(len(metro_indices)):
    #Get metro index in dataframe, figure out where to stop looking for the rate
    index = metro_indices[x]
    next_index = metro_indices[x + 1] if not x == (len(metro_indices) - 1) else len(crime)
    
    #Search the bounded range for the rate
    for rate_index in range(index, next_index):
        row = crime.iloc[rate_index, :]
        if row['Counties/principal cities'] == 'Rate per 100,000 inhabitants' and not np.isnan(row['Violent\ncrime']) and not np.isnan(row['Property\ncrime']):
            final_metro_indices.append(index)
            violent_crime_rate.append(row['Violent\ncrime'])
            property_crime_rate.append(row['Property\ncrime'])
            break

#Sanity check on the number of values for indices and crime rates
assert(len(final_metro_indices) == len(violent_crime_rate))
assert(len(violent_crime_rate) == len(property_crime_rate))

#Strip down the metro names so that we can match against the merged dataframe
metro_names = []
for index in final_metro_indices:
    name = crime.iloc[index,:]['Metropolitan Statistical Area']
    name = name.split(' M.S.A.')[0]
    metro_names.append(name)

#Build the dataframe for 
crime_df = pd.DataFrame(data={
    '2013 CBSA' : metro_names,
    'Violent crime rate per 100,000' : violent_crime_rate,
    'Property crime rate per 100,000' : property_crime_rate
})
```

```python
merged = merged.merge(crime_df, on='2013 CBSA', how='inner')
```
